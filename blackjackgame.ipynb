```python
pic = 10
cardpoints = { 'ONE':'one', 'TWO':2, 'THREE':3, 'FOUR':4, 'FIVE':5, 
         'SIX':6, 'SEVEN':7, 'EIGHT':8, 'NINE':9, 'TEN':10, 
         'JACK':pic, 'QUEEN':pic, 'KING':pic  }

cards = ['ONE', 'TWO', 'THREE', 'FOUR', 'FIVE', 
         'SIX', 'SEVEN', 'EIGHT', 'NINE', 'TEN', 
         'JACK', 'QUEEN', 'KING']
```

```python
class CardDraw:
    
    def __init__(self):
        self.playcards = cards*4
        self.drawtime = 1
        
    def draw(self):
        self.num = random.randint(0,53-self.drawtime)
        self.drawtime += 1
        self.card = self.playcards.pop(self.num)
        self.point = cardpoints[self.card]
        return [self.card, self.point]
    
    def rest(self):
        print(self.playcards)
```

```python
class GamePlay(CardDraw):
    
    def __init__(self):
        self.play = CardDraw()
        
    
    def firstsets(self):
        self.player_card1 = self.play.draw()
        self.player_card2 = self.play.draw()
        self.dealer_card1 = self.play.draw()
        self.dealer_card2 = self.play.draw()
        
        if self.dealer_card1[0] and self.dealer_card2[0] == 'ONE':
            self.dealer_card1[1] = 10
            self.dealer_card2[1] = 10
            
            return [self.player_card1, self.player_card2, self.dealer_card1, self.dealer_card2]
            
        elif self.dealer_card1[0] == 'ONE':
            if self.dealer_card2[1] < 8:
                self.dealer_card3 = self.play.draw()
                return [self.player_card1, self.player_card2, self.dealer_card1, 
                        self.dealer_card2, self.dealer_card3]
            else:
                return [self.player_card1, self.player_card2, self.dealer_card1, self.dealer_card2]
            
        elif self.dealer_card2[0] == 'ONE':
            if self.dealer_card1[1] < 8:
                self.dealer_card3 = self.play.draw()
                return [self.player_card1, self.player_card2, self.dealer_card1, 
                        self.dealer_card2, self.dealer_card3]
            else:
                return [self.player_card1, self.player_card2, self.dealer_card1, self.dealer_card2]
            
        elif self.dealer_card1[1] + self.dealer_card2[1] < 15:
            self.dealer_card3 = self.play.draw()
            return [self.player_card1, self.player_card2, self.dealer_card1, 
                        self.dealer_card2, self.dealer_card3]
        else:
            return [self.player_card1, self.player_card2, self.dealer_card1, self.dealer_card2]
        
        
    
    def resetcards(self):
        self.playcards = cards*4
        self.drawtime = 1
        
            
    
    
    def choiceone(self):
        self.bbb = True
        
        while self.bbb:
            try:      
                self.answerone = int(input('Which whould you like to set card ONE? 1 or 10: '))
                self.choice = [1, 10]
                
                if not self.answerone in self.choice:
                    print("Something went wrong. The answer must be either 1 or 10.")
                    pass
                else:
                    if self.answerone ==1:
                        self.bbb = False
                        return 1
                    else:
                        self.bbb = False
                        return 10
            except:
                print("Something went wrong. The answer must be either 1 or 10.")
                pass
    
    
    
    def choicedraw(self):
        self.aaa = True
        while self.aaa:
            
            try:
                self.answer = input('Would you like to drew another card? YES or NO: ')
                self.choice = ['yes','no']
                if not self.answer.lower() in self.choice:
                    print("Something went wrong. The answer must be either 'YES' or 'NO'.")
                    pass

                else:
                    if self.answer.lower() == 'yes':
                        self.player_card3 = self.play.draw()
                        self.aaa = False
                        return self.player_card3
                    else:
                        self.aaa = False
                        return 0
            except :
                print("Something went wrong. The answer must be either 'YES' or 'NO'.")
                pass

            
```
